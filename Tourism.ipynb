```python
numerical_columns = df_data.select_dtypes(include=['int64', 'float64'])
num_cols = len(numerical_columns.columns)
num_rows = (num_cols + 2) // 3
plt.figure(figsize=(10, 2 * num_rows))
for i, column in enumerate(numerical_columns.columns, 1):
    plt.subplot(num_rows, 3, i)
    sns.histplot(data=df_data, x=column, kde=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()
```

```python
num_cols = df_data.select_dtypes(include=['int64', 'float64']).columns
for col in num_cols:
  plt.figure(figsize=(10, 6))
  plt.subplot(1, 2, 1)
  sns.histplot(df_data[col], kde=True)
  plt.title(f'Distribution of {col}')
  plt.subplot(1, 2, 2)
  sns.boxplot(x=df_data[col])
  plt.title(f'Boxplot of {col}')
  plt.tight_layout()
  plt.show()
```

```python
numerical_columns = df_data.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(10, 6))
sns.boxplot(numerical_columns.corr())
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='unemployment', y='tourism_receipts', data=df_data)
plt.title('Unemployment vs Tourism Receipts')
plt.xlabel('Unemployment')
plt.show()
```

```python
for col in df_data.columns:
  if col != 'unemployment':
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='unemployment', y=col, data=df_data)
    plt.title(f'{col} vs Unemployment')
    plt.xlabel('Unemployment')
    plt.ylabel(col)
    plt.show()
```

```python
grouped_states = df_data.groupby('unemployment')[df_data.select_dtypes(include=['int64', 'float64']).columns].mean()
# Define numerical_cols within this cell to ensure it's accessible
grouped_states.plot(kind='bar', figsize=(10, 6))
plt.title('Mean Values of Numerical Columns by State')
plt.xlabel('Feature')
plt.ylabel('Mean Value')
plt.xticks(rotation=45)
plt.show()
```
